# 1. 1D linear convection

$$\Large 
                \frac{\delta u}{\delta t}+c\frac{\delta u}{\delta x} =0
$$


Ecuacion discreta

$$\Large 
                \frac{u_i^{n+1}-u_i^n}{\Delta t}+c\frac{u_i^n-u^n_{i-1}}{\Delta x}=0
$$

$$\Large 
                u_i^{n+1}=u_i^n-c\frac{\Delta t}{\Delta x}(u_i^n-u^n_{i-1})=0
$$

Condiciones iniciales

cuando 0-5 <= x <= 1 -> u=2

sino u = 1 en en los demas valores dentro del intervalo (0,2)

Condiciones de frontera

u=1  en x=0,2

```python
import numpy as np
import matplotlib.pylab as plt
%config Completer.use_jedi = False #for auto complete code 
```

```python
nx=20
nt=50
dt=0.01
dx=2/(nx-1)
c=1
#mesh
x=np.linspace(0,2,20)
#variable u
u=np.zeros(len(x))
```

```python
# Condiciones de frontera
for i in range(nx):
    if x[i]>=0.5 and x[i]<=1:
        u[i]=2
    else:
        u[i]=1
```

```python
plt.figure(figsize=(15,10))
plt.plot(x,u)
plt.title('Condiciones iniciales',fontsize=30)
plt.show()
```

```python
#diferencias finitas

for t in range(nt):
    u_n=u
    for i in range(1,nx-1):
        u[i]=u_n[i]-c*(dt/dx)*(u_n[i]-u_n[i-1])
```

```python
plt.figure(figsize=(15,10))
plt.plot(u)
plt.title('U despues de '+str(dt*nt)+' segundos',fontsize=30)
plt.show()
```

# 2. 1 D Convection, " Flujo no viscoso burger"

$$\Large 
                \frac{\delta u}{\delta t}+u\frac{\delta u}{\delta x} =0
$$

$$\Large 
                u_i^{n+1}=u_i^n+u_i^n\frac{\Delta t}{\Delta x}(u_i^n-u^n_{i-1})=0
$$

```python
nx=20
nt=50
dt=0.01
dx=2/(nx-1)
#mesh
x=np.linspace(0,2,20)
#variable u
u=np.zeros(len(x))
```

```python
# Condiciones de frontera
for i in range(nx):
    if x[i]>=0.5 and x[i]<=1:
        u[i]=2
    else:
        u[i]=1
```

```python
#diferencias finitas

for t in range(nt):
    u_n=u
    for i in range(1,nx-1):
        u[i]=u_n[i]-u_n[i]*(dt/dx)*(u_n[i]-u_n[i-1])
```

```python
plt.figure(figsize=(15,10))
plt.plot(u)
plt.title('U despues de '+str(dt*nt)+' segundos',fontsize=30)
plt.show()
```

# 3. 1D diffusion

```python
nx=20
nt=50
dt=0.01
dx=2/(nx-1)
vis=0.1
#mesh
x=np.linspace(0,2,20)
#variable u
u=np.zeros(len(x))
```

```python
# Condiciones de frontera
for i in range(nx):
    if x[i]>=0.5 and x[i]<=1:
        u[i]=2
    else:
        u[i]=1
```

```python
#diferencias finitas

for t in range(nt):
    u_n=u
    for i in range(1,nx-1):
        u[i]=u_n[i]+vis*(dt/dx**2)*(u_n[i+1]-2*u_n[i]+u_n[i-1])
```

```python
plt.figure(figsize=(15,10))
plt.plot(u)
plt.title('U despues de '+str(dt*nt)+' segundos',fontsize=30)
plt.show()
```

# 4. 1 D Burger's  equation

$$\Large 
                \frac{\delta u}{\delta t}+u\frac{\delta u}{\delta x} = \nu \frac{\delta^2 u }{\delta^2 x }
$$

Condiciones iniciales

$$\Large 
                 u= -2\nu \frac{\frac{\delta \phi}{\delta x}}{\phi}+4
$$

$$\Large 
                \phi = e^{\frac{-x^2}{4\nu}} + e^{\frac{-(x-2\pi)^2}{4\nu}}
$$

Condiciones de frontera

$ u(0)= u(2\pi) $

Solución analitica


$$\Large 
                 u= -2\nu \frac{\frac{\delta \phi}{\delta x}}{\phi}+4
$$

$$\Large 
                \phi = e^{\frac{-(x-4t)^2}{4\nu(t+1)}} + e^{\frac{-(x-4t-2\pi)^2}{4\nu(t+1)}}
$$
